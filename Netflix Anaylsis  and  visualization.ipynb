```python
# dropping the show_id columns cause it not needed
df.drop(columns=["show_id"], inplace = True)
```

```python
# dropping the description columns because it not needed 
df.drop(columns=["description"], inplace= True)
```

```python
df
```

```python
# Renaming some columns that are not name properly
df.rename(columns={"type": "Content type","country":"Production_country","listed_in":"Genres"}, inplace = True)
df
```

```python
# resetting my index after drop some missing values , it was not align properly , so i reset it
df.reset_index(drop= True, inplace= True)
```

```python
df
```

```python
# Grouping the rating columns into four categories cause is clear the other way
conditions = [
    df['rating'].isin(['TV-Y', 'TV-Y7', 'TV-Y7-FV', 'TV-G', 'G']),
    df['rating'].isin(['TV-PG', 'PG', 'PG-13', 'TV-14']),
    df['rating'].isin(['TV-MA', 'R', 'NC-17']),
    df['rating'].isin(['NR', 'UR'])
]
choices = ['Kids', 'Teen', 'Mature', 'Unrated']


df['rating_category'] = np.select(conditions, choices, default='Unrated')

```

```python
df
```

```python
# droping the rating columns cause i created the new rating category which is more clear 
df.drop(columns='rating', inplace = True)
```

```python
df["Production_country"].nunique()
```

```python
# i convert my date added to datetime format
import datetime 
from datetime import datetime, time , timedelta, date

df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce').dt.date
df
```

```python
df.isnull().sum()
```

```python
df.dropna(subset=['date_added'], inplace =True)
```

```python
df.isnull().sum()
```

```python
df["date_added"].value_counts()
```

```python
# i split the duration to duration value and duration unit 
# Extract the number and the unit (min/season)
duration_split = df['duration'].str.extract(r'(\d+)\s*(\w+)')

# Assign to new columns
df['duration_value'] = pd.to_numeric(duration_split[0])
df['duration_unit'] = duration_split[1].str.lower()

# Standardize plurals: seasons -> season, mins -> min
df['duration_unit'] = df['duration_unit'].replace({
    'mins': 'min',
    'minutes': 'min',
    'minute': 'min',
    'seasons': 'season'
})
```

```python
df
```

```python
# so after splitting my duration, i drop the main duration column
df.drop(columns="duration", inplace = True)
```

```python
# Rearranging my columns to the way i want it to be
df[["Content type","title","cast","Production_country","date_added","release_year","rating_category","duration_value","duration_unit","Genres"]]
```

```python
from collections import Counter
cast_series = df['cast']

# Split by comma into lists of names
all_cast = cast_series.str.split(', ')

# Flatten into a single list of names
flat_cast_list = [actor for sublist in all_cast for actor in sublist]

# Count occurrences
cast_counts = Counter(flat_cast_list)

# Get the top most common
most_common_cast = cast_counts.most_common(10)

```

```python
# Group by year and genre, count occurrences
genre_counts = df.groupby(['release_year', 'Genres']).size().reset_index(name='Count')

# For each year, get the genre with the highest count
top_genre_per_year = genre_counts.loc[genre_counts.groupby('release_year')['Count'].idxmax()]

print(top_genre_per_year)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
plt.scatter(top_genre_per_year['release_year'], top_genre_per_year['Count'], color='teal')
plt.title('Most Released Genre by Year (Number of Titles)')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles in Top Genre')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Group by Release_Year and Rating_Category
rating_counts = df.groupby(['release_year', 'rating_category']).size().reset_index(name='Count')

# Pivot to make Rating_Category become columns
rating_pivot = rating_counts.pivot(index='release_year', columns='rating_category', values='Count').fillna(0)

# Plot
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))
rating_pivot.plot(kind='line', figsize=(12,6))
plt.title('Number of Movies Released Over Years by Rating Category')
plt.xlabel('Release Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Filter only movies and ensure we have duration + genre
movies = df[(df['Content type'] == 'Movie') & df['duration_value'].notna() & df['Genres'].notna()]

# Group by genre, compute average duration
avg_duration_by_genre = movies.groupby('Genres')['duration_value'].mean().sort_values(ascending=False)



plt.figure(figsize=(10,6))
avg_duration_by_genre.head(10).plot(kind='barh', color='slateblue')
plt.title('Top 10 Genres with Highest Average Movie Duration')
plt.xlabel('Average Duration (minutes)')
plt.gca().invert_yaxis()
plt.tight_layout()
plt.show()
```
